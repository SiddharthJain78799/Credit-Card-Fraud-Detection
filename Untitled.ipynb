```python
from tkinter import *
import joblib

# Predict Function
def predict():
    values = [float(e.get()) for e in entries]
    model = joblib.load('model_credit.pkl')
    y_pred = model.predict([values])[0]
    
    result = "Normal Transaction" if y_pred == 0 else "Fraudulent Transaction"
    Label(root, text=f"Prediction: {result}", fg="blue").grid(row=31, columnspan=2)

# Tkinter Window
root = Tk()
root.title("Credit Card Fraud Detection System")

Label(root, text="Credit Card Fraud Detection System", bg="black", fg="white", width=40).grid(row=0, columnspan=2)

entries = []
for i in range(29):
    Label(root, text=f"Enter value of V{i+1}").grid(row=i+1, column=0)
    e = Entry(root)
    e.grid(row=i+1, column=1)
    entries.append(e)

Button(root, text='Predict', command=predict).grid(row=30, column=1, pady=4)

root.mainloop()
```
